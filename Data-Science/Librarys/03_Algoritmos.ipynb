# **Qué es un algoritmo?**

Un algoritmo es un conjunto de instrucciones o reglas definidas y ordenadas que se utilizan para resolver un problema específico o realizar una tarea. 

Los algoritmos son fundamentales en la programación y la informática, ya que proporcionan una forma sistemática de abordar problemas y encontrar soluciones.

---

# **Características de un algoritmo**

1. **Finitud**: Un algoritmo debe terminar después de un número finito de pasos.

2. **Definición**: Cada paso del algoritmo debe estar claramente definido y ser preciso.

3. **Entrada**: Un algoritmo puede tener cero o más entradas, que son los datos necesarios para llevar a cabo el proceso.

4. **Salida**: Un algoritmo produce al menos una salida, que es el resultado del proceso.

5. **Efectividad**: Los pasos del algoritmo deben ser lo suficientemente efectivos como para ser llevados a cabo en un tiempo razonable.

6. **Generalidad**: Un algoritmo debe ser aplicable a una clase de problemas, no solo a un caso específico.

---

# **Ejemplo de un algoritmo simple**

## **Algoritmo para sumar dos números**

1. **Inicio**

2. **Entrada**: Leer dos números, A y B.

3. **Proceso**: Calcular la suma de A y B.

4. **Salida**: Mostrar el resultado de la suma.

5. **Fin**

## **Representación del algoritmo**

```plaintext
Inicio
    Leer A
    Leer B
    Suma = A + B
    Mostrar Suma
Fin
```

# **Importancia de los algoritmos**

Los algoritmos son esenciales en la programación y la informática porque:

- Proporcionan una estructura clara para resolver problemas.

- Permiten la automatización de tareas repetitivas.

- Facilitan la comprensión y el mantenimiento del código.

- Son la base para el desarrollo de software eficiente y efectivo.
