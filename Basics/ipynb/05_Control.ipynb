````python
# Iterar sobre una lista
frutas = ["manzana", "pera", "uva"]

for fruta in frutas:
    print("Fruta:", fruta)

# Iterar usando `range()`
for i in range(5):  # 0, 1, 2, 3, 4
    print("N煤mero:", i)

# Iterar con 铆ndice y valor usando `enumerate()`
for i, fruta in enumerate(frutas):
    print(f"ndice {i}: {fruta}")
````
